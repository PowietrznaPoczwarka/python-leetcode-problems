```python
# Merge Two Sorted Lists
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        headlist = []

        while True:
            # if first value from list1 is smaller, take it, else take the other one
            if list1 != None and list2 != None:
                if list1.val < list2.val:
                    headlist.append(list1.val)
                    list1 = list1.next
                else:
                    headlist.append(list2.val)
                    list2 = list2.next
            elif list1 == None:
                head = list2
                break
            else:
                head = list1
                break

        headlist.reverse()
        for i in range(len(headlist)):
            head = ListNode(val = headlist[i], next = head)
        
        return head

Solution().mergeTwoLists(ListNode(1, ListNode(2, ListNode(4))), ListNode(1, ListNode(3, ListNode(4))))
```

```python
# Generate Parentheses
from typing import List

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        output = [""]
        options = ['(', ')']

        def expand(inputs_list):
            list_to_extend = []
            for option in options:
                list_to_extend.extend([s + option for s in inputs_list])
            # print(f"Koniec: {list_to_extend}")
            return list_to_extend
        
        def invalidate(s):
            string2 = ''
            for char in s:
                if char == "(":
                    string2 += char
                elif char == ")":
                    if len(string2)>0 and string2[-1] == "(":
                        string2 = string2[:-1]
                    else:
                        return True
            if len(string2) != 0:
                return True
            
            return False

        for i in range(2*n):
            print(i)
            output = expand(output)

        # print(output,"\n")
        for solution in output[:]:
            # print("solution", solution)
            if invalidate(solution):
                # print("inwalidujemy", solution, "\n")
                output.remove(solution)

        return output
    
Solution().generateParenthesis(3)
```

```python
# Merge k Sorted Lists

from typing import List, Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        temp = []

        for list in lists:
            if not isinstance(list, ListNode):
                continue

            l = list
            while l != None:
                temp.append(l.val)
                l = l.next

        temp.sort()
            
        if len(temp) != 0:
            result = ListNode(temp[0])
            current = result

            for value in temp[1:]:
                current.next = ListNode(value)
                current= current.next

            return result
        else:
            return None
        
Solution().mergeKLists([ListNode(1, ListNode(4, ListNode(5))), ListNode(1, ListNode(3, ListNode(4))), ListNode(2, ListNode(6))])
```

```python
# Swap Nodes in Pairs

from typing import Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        previous = None
        current = head

        while current and current.next:
            first = current
            second = current.next

            first.next = second.next
            second.next = first

            if previous:
                previous.next = second
                
            else:
                head = second
                print("first run")

            previous = first
            current = first.next

        return head
    
Solution().swapPairs(ListNode(1, ListNode(2, ListNode(3, ListNode(4)))))
```

```python
# Remove Duplicates from Sorted Array

from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return 0
        
        current = nums[0]

        k = 1
        for el in nums[1:]:
            # print(el, current)
            if el != current:
                k += 1
                nums[k-1] = el

            current = el
        return k

Solution().removeDuplicates([0,0,1,1,1,2,2,3,3,4])
```

```python
# Remove Element
from typing import List

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        count = 0
        
        for i in range(len(nums)):
            if nums[i] != val:
                nums[count] = nums[i]
                count += 1
        
        # print(nums)
        return count
    
Solution().removeElement([3,2,2,3], 3)
```

```python
# Find Index of First Occurrence in a String

class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if len(needle) == 0:
            return 0
        
        for i in range(len(haystack) - len(needle)+1):
            # print(i)
            # print(haystack[i:(i+len(needle))])
            
            if haystack[i:(i+len(needle))] == needle:
                # print(haystack[i:(i+len(needle))], needle)
                return i
            
        return -1

Solution().strStr("sadbutsad", "sad")
```

```python
class Solution:
    def divide(self, dividend: int, divisor: int) -> int:
        
        min = -2147483648
        max = 2147483647
        
        if dividend == min and divisor == -1:
            return max
        
        result = 0
        
        sign = -1 if (dividend < 0) ^ (divisor < 0) else 1 # XOR, false if both true or both false, else true
        
        dividend, divisor = abs(dividend), abs(divisor)
        
        # while dividend >= divisor:
        #     dividend -= divisor
            
        #     result += 1
            
        # return result*sign
                    
        while dividend >= divisor:
            temp = divisor
            multiple = 1

            # inner loop
            # if temp*2 is less than divident, double result and temp
            while dividend >= temp * 2:
                temp *= 2
                multiple *= 2

            dividend -= temp
            result += multiple
        
        return result * sign
    
Solution().divide(10, -3)
```
