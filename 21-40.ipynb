```python
# Merge Two Sorted Lists
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        headlist = []

        while True:
            # if first value from list1 is smaller, take it, else take the other one
            if list1 != None and list2 != None:
                if list1.val < list2.val:
                    headlist.append(list1.val)
                    list1 = list1.next
                else:
                    headlist.append(list2.val)
                    list2 = list2.next
            elif list1 == None:
                head = list2
                break
            else:
                head = list1
                break

        headlist.reverse()
        for i in range(len(headlist)):
            head = ListNode(val = headlist[i], next = head)
        
        return head

Solution().mergeTwoLists(ListNode(1, ListNode(2, ListNode(4))), ListNode(1, ListNode(3, ListNode(4))))
```

```python
# Generate Parentheses
from typing import List

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        output = [""]
        options = ['(', ')']

        def expand(inputs_list):
            list_to_extend = []
            for option in options:
                list_to_extend.extend([s + option for s in inputs_list])
            # print(f"Koniec: {list_to_extend}")
            return list_to_extend
        
        def invalidate(s):
            string2 = ''
            for char in s:
                if char == "(":
                    string2 += char
                elif char == ")":
                    if len(string2)>0 and string2[-1] == "(":
                        string2 = string2[:-1]
                    else:
                        return True
            if len(string2) != 0:
                return True
            
            return False

        for i in range(2*n):
            print(i)
            output = expand(output)

        # print(output,"\n")
        for solution in output[:]:
            # print("solution", solution)
            if invalidate(solution):
                # print("inwalidujemy", solution, "\n")
                output.remove(solution)

        return output
    
Solution().generateParenthesis(3)
```

```python
# Merge k Sorted Lists

from typing import List, Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        temp = []

        for list in lists:
            if not isinstance(list, ListNode):
                continue

            l = list
            while l != None:
                temp.append(l.val)
                l = l.next

        temp.sort()
            
        if len(temp) != 0:
            result = ListNode(temp[0])
            current = result

            for value in temp[1:]:
                current.next = ListNode(value)
                current= current.next

            return result
        else:
            return None
        
Solution().mergeKLists([ListNode(1, ListNode(4, ListNode(5))), ListNode(1, ListNode(3, ListNode(4))), ListNode(2, ListNode(6))])
```

```python
# Remove Duplicates from Sorted Array

from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return 0
        
        current = nums[0]

        k = 1
        for el in nums[1:]:
            # print(el, current)
            if el != current:
                k += 1
                nums[k-1] = el

            current = el
        return k

Solution().removeDuplicates([0,0,1,1,1,2,2,3,3,4])
```

```python
# Remove Element
from typing import List

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        count = 0
        
        for i in range(len(nums)):
            if nums[i] != val:
                nums[count] = nums[i]
                count += 1
        
        # print(nums)
        return count
    
Solution().removeElement([3,2,2,3], 3)
```

```python
class Solution:
    def divide(self, dividend: int, divisor: int) -> int:
        
        min = -2147483648
        max = 2147483647
        
        if dividend == min and divisor == -1:
            return max
        
        result = 0
        
        sign = -1 if (dividend < 0) ^ (divisor < 0) else 1 # XOR, false if both true or both false, else true
        
        dividend, divisor = abs(dividend), abs(divisor)
        
        # while dividend >= divisor:
        #     dividend -= divisor
            
        #     result += 1
            
        # return result*sign
                    
        while dividend >= divisor:
            temp = divisor
            multiple = 1

            # inner loop
            # if temp*2 is less than divident, double result and temp
            while dividend >= temp * 2:
                temp *= 2
                multiple *= 2

            dividend -= temp
            result += multiple
        
        return result * sign
    
Solution().divide(10, -3)
```

```python
class Solution:
    def longestValidParentheses(self, s: str) -> int:
        # print("input string:", s)
        
        left = 0
        right = 0
        max_length = 0
        
        for char in s:
            if char == "(":
                left += 1
            elif char == ")":
                right += 1
                
            # if at any point there are more closing than opening, the substring so far is invalid -> reset counters
            if right > left:
                left = 0
                right = 0
                
            # if at any point left == right -> valid substring, update max_length if more than previous max_length
            
            if left == right:
                max_length = max(max_length, left + right)
                
        print("max_length pass 1:", max_length)
        
        left = 0
        right = 0
        
        reversed_s = s[::-1]
        for char in reversed_s:
            if char == "(":
                left += 1
            elif char == ")":
                right += 1
                
            # if at any point there are more opening than closing, the substring so far is invalid -> reset counters
            if left > right:
                left = 0
                right = 0
                
            # if at any point left == right -> valid substring, update max_length if more than previous max_length
            
            if left == right:
                max_length = max(max_length, left + right)
                
        print("max_length pass 2:", max_length)
        
        return max_length
        
Solution().longestValidParentheses("(()")
```

```python
# Find first and last position of element in sorted array

class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        # nums sorted in non decreasing order.
        
        first = -1
        for i in range(len(nums)):
            if nums[i] == target:
                first = i
                break
            
        if first == -1:
            return [-1, -1]
        else:
            last = first
            for j in range(first, len(nums)):
                if nums[j] == target:
                    last = j
                    
                    if len(nums) > j+1:
                        if nums[j+1] != target:
                            break
                    
            return [first, last] 

Solution().searchRange([5,7,7,8,8,10], 8)
```

```python
class Solution:
    def solveSudoku(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        
        # look up tables initialization
        # lists of 9 lets...
        rows = [set() for _ in range(9)]
        cols = [set() for _ in range(9)]
        boxes = [set() for _ in range(9)]
        
        
        for r in range(9):
            for c in range(9):
                if board[r][c] != ".":
                    val = board[r][c]
                    rows[r].add(val)
                    cols[c].add(val)
                    
                    # determine box id
                    box_id = (r // 3) * 3 + (c // 3)
                    boxes[box_id].add(val)

        print("Initial rows:", rows)
        print("Initial cols:", cols)
        print("Initial boxes:", boxes)

        def solve(r, c):
            # if moved beyond last row, it means it is solved
            if r == 9:
                return True
            
            # next cell coordinates
            next_r, next_c = (r, c + 1) if c < 8 else (r + 1, 0)
            
            # skip filled cells
            if board[r][c] != ".":
                return solve(next_r, next_c)
            
            
            # if not, tru solving for this cell
            
            box_id = (r // 3) * 3 + (c // 3)
            
            for char in "123456789":
                # check validity
                if char not in rows[r] and char not in cols[c] and char not in boxes[box_id]:
                    
                    # if valid, place number and update cells
                    board[r][c] = char
                    rows[r].add(char)
                    cols[c].add(char)
                    boxes[box_id].add(char)
                    
                    # move to next cell
                    if solve(next_r, next_c):
                        return True
                    
                    # if solving fails, backtrack
                    board[r][c] = "."
                    rows[r].remove(char)
                    cols[c].remove(char)
                    boxes[box_id].remove(char)
            
            return False

        # start solving from top-left cell
        solve(0, 0)
        
board_2 = [[".",".",".",".",".",".",".",".","."],[".","9",".",".","1",".",".","3","."],[".",".","6",".","2",".","7",".","."],[".",".",".","3",".","4",".",".","."],["2","1",".",".",".",".",".","9","8"],[".",".",".",".",".",".",".",".","."],[".",".","2","5",".","6","4",".","."],[".","8",".",".",".",".",".","1","."],[".",".",".",".",".",".",".",".","."]]                        
Solution().solveSudoku(board_2)
```

```python
# 38. Count and Say

# Run-length encoding (RLE) is a string compression method that works 
# by replacing consecutive identical characters (repeated 2 or more times) 
# with the concatenation of the character and the number marking the 
# count of the characters (length of the run). For example, to compress 
# the string "3322251" we replace "33" with "23", replace "222" with "32", 
# replace "5" with "15" and replace "1" with "11". 
# Thus the compressed string becomes "23321511".

# 3322251 -> 2322251 -> 233251 -> 2332151 -> 23321511

# countAndSay(4) is the RLE of countAndSay(3)
# countAndSay(3) is the RLE of countAndSay(2)
# countAndSay(2) is the RLE of countAndSay(1) which is 1

# RLE of 1 is 11; (countAndSay(2) = "11")
# RLE of 11 is 21; (countAndSay(3) = "21")
# RLE of 21 is 1211; (countAndSay(4) = "1211")

# 1 <= n <= 30

class Solution:
    def rle(self, s: str) -> str:
        result = []

        prev_char = s[0]
        counter = 0

        for char in s:
            if char == prev_char:
                counter += 1
                # all good, move further.
            else:
                # print(counter * prev_char)
                # result += str(counter) + prev_char
                result.append(str(counter))
                result.append(prev_char)
                # reset the counter, previous character changes
                counter = 1
                prev_char = char
        
        # reached the end of the string, append what's left
        result.append(str(counter))
        result.append(prev_char)
        return "".join(result)


    def countAndSay(self, n: int) -> str:
        if n == 1:
            return "1"
        
        else:
            return self.rle(self.countAndSay(n-1))
        
Solution().countAndSay(4)
```

```python
# 39. Combination Sum
# we look for all unique combinations in candiates where they sum to target. with repetition.
from typing import List

class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:

        # we need temporary function that we can use recusively. 
        # it would take in the same list of candidates (candidates) and some input (like how far away current sum is from the target)
        # and simply append one of the candidates to the list
        # also it would have a check if current sum equals target (in this case its returned) or went over the target (returns None)

        # main function would simply append all the results from this function?

        results = []

        def backtrack(current, remaining):

            # hit the target
            if remaining == 0:
                results.append(current)
                return  # Tweak 2: Added return here so it stops digging

            # over the target
            elif remaining < 0:
                return

            # loop through candidates and dive deeper
            for c in candidates:
                temp = current[:]
                temp.append(c)
                backtrack(current=temp, remaining=remaining - c)
                
        # Tweak 1: Pass the initial target in
        backtrack([], target)

        # now get rid of duplicates:
        uniques = set()
        for res in results:
            combination = tuple(sorted(res))
            uniques.add(combination)

        final_results = [list(item) for item in uniques]

        return final_results
    
Solution().combinationSum(candidates = [2,3,6,7], target = 7)
```

```python
# Given a collection of candidate numbers (candidates) and a target number (target), find all unique combinations in candidates where the candidate numbers sum to target.
# Each number in candidates may only be used once in the combination.
# Note: The solution set must not contain duplicate combinations.
# ie. same as previous problem but no repetition allowed

class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:

        # we need temporary function that we can use recusively. 
        # it would take in the list of trimmed candidates and some input (like how far away current sum is from the target)
        # and simply append one of the (remaining) candidates to the list
        # also it would have a check if current sum equals target (in this case its returned) or went over the target (returns None)

        # main function would simply append all the results from this function?
        
        results = []

        def backtrack(current, remaining, remaining_candidates):

            # hit the target
            if remaining == 0:
                results.append(current)
                return  # Tweak 2: Added return here so it stops digging

            # over the target
            elif remaining < 0:
                return

            # loop through candidates and dive deeper
            for i, c in enumerate(remaining_candidates):
                temp = current[:]
                temp.append(c)
                remaining_after_c = remaining_candidates[:]
                remaining_after_c.pop(i)
                backtrack(current=temp, remaining=remaining - c, remaining_candidates=remaining_after_c)

        backtrack([], target, candidates)

        # print(results)
        # now get rid of duplicates:
        uniques = set()
        for res in results:
            combination = tuple(sorted(res))
            uniques.add(combination)
        
        final_results = [list(item) for item in uniques]
        
        # print("final results:", final_results)
        return final_results
    
Solution().combinationSum2(candidates = [10,1,2,7,6,1,5], target = 8)
```

```python
# better approach...
# first we can sort the candidates - skip numbers greater than target
# this way we can prevent duplicate search (cause now, given a list with two 1s, we are creating the same exact tree twice)

class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:

        candidates.sort()   
        results = []

        def backtrack(current, remaining, start_index):

            # hit the target
            if remaining == 0:
                results.append(current)
                return

            # over the target
            elif remaining < 0:
                return

            # loop through candidates and dive deeper
            # for i, c in enumerate(remaining_candidates):
            for i in range(start_index, len(candidates)):
                c = candidates[i]
                if i > start_index and c == candidates[i - 1]: 
                    # we already tried this number at this level - because current candidate is the same as the previous
                    # if i is greater than start_index, it means we are at the same level of the tree (denoted by length of current list)
                    # it would trigger at the second 1 in the sorted list of candidates
                    continue

                backtrack(current=current + [c], remaining=remaining - c, start_index=i + 1)

        backtrack([], target, 0)

        return results

    
Solution().combinationSum2(candidates = [10,1,2,7,6,1,5], target = 8)
```
