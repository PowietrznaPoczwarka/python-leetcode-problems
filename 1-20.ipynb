```python
# Two Sum

from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(len(nums)):
                if i != j and nums[i] + nums[j] == target:
                    return [i, j]
                
Solution().twoSum([2,7,11,15], 9)
```

```python
# Add Two Numbers

from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        l1_str = ''
        l2_str = ''

        while l1 is not None:
            l1_str += str(l1.val)
            l1 = l1.next
        
        while l2 is not None:
            l2_str += str(l2.val)
            l2 = l2.next

        l3_str = str(int(l1_str[::-1]) + int(l2_str[::-1])) # to nie jest odwrocone

        l3 = None
        i = 0
        while i < len(l3_str):
            l3 = ListNode(int(l3_str[i]), l3)
            i += 1

        return l3
    
x = Solution().addTwoNumbers(ListNode(2, ListNode(4, ListNode(3))), ListNode(5, ListNode(6, ListNode(4))))

print(x.val, x.next.val, x.next.next.val)
```

```python
# Longest Substring Without Repeating Characters

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        
        if len(s) == 0:
            return 0

        substring = s[0]
        max_substring = s[0]
        i = 1
        
        while i <= len(s) - 1:
            if s[i] not in substring:
                substring += s[i]
            else:
                substring = substring[substring.find(s[i])+1::] + s[i]
                #substring = s[i]
            i += 1
            
            if len(substring) > len(max_substring):
                max_substring = substring
        
        return len(max_substring)
    
Solution().lengthOfLongestSubstring("abcabcbb")
```

```python
# Median of Two Sorted Arrays

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        nums3 = nums1 + nums2
        nums3.sort()
        l = len(nums3)

        if l % 2 == 0:
            #parzyste
            return (nums3[int(l/2 - 1)]+nums3[int(l/2)])/2

        else:
            return nums3[int(l/2 - 0.5)]
        
Solution().findMedianSortedArrays([1,3], [2])
```

```python
# Longest Palindormic Substring

class Solution:
    def longestPalindrome(self, s: str) -> str:
        max_palindrome = s[0]

        for i,c in enumerate(s):

            substring = c
            m = i #start of the palindrome
            n = i #end of the palindrome

            #starting palindrome
            while n+1 <= len(s)-1 and s[n+1] == s[i]: #if possible and makes sense
                substring += s[n+1]
                n += 1

            #expand palindrome
            while n+1 <= len(s)-1 and m-1 >= 0 and s[n+1] == s[m-1]: #if possible and makes sense
                substring = s[m-1]+ substring + s[n+1] #add at the start and end of the palindrome
                n += 1 #change end pos. of the palindrome
                m -= 1 #change start pos. of the palindrome

            if len(max_palindrome) <= len(substring): #check if combined palindrome can be the longest one
                max_palindrome = substring

        return max_palindrome
    
Solution().longestPalindrome("babad")
```

```python
# ZigZag Conversion

class Solution:
    def convert(self, s: str, numRows: int) -> str:
        step1 = max(numRows*2 - 2,1)
        string = ""
        
        for i in range(numRows):
            if i == 0 or i == numRows-1:
                substring = s[i::step1]
            else:
                substring_ver = {j: s[j] for j in range(i, len(s), step1)}
                substring_dia = {j: s[j] for j in range(step1-i, len(s), step1)}
                substring_dict = dict(sorted({**substring_ver, **substring_dia}.items()))
                substring = ''.join(dict(sorted(substring_dict.items())).values())
                
            string += substring
            
        return string
    
Solution().convert("PAYPALISHIRING", 3)
```

```python
# Reverse Integer

class Solution:
    def reverse(self, x: int) -> int:
        x_str = str(x)
        if x_str[0] == '-':
            x_rev = int('-' + str(x)[:0:-1])
        else:
            x_rev = int(str(x)[::-1])
            
        if x_rev > 2147483648 or x_rev < -2147483648:
            x_rev = 0
        return x_rev

Solution().reverse(-123)
```

```python
# String to Integer (atoi)

class Solution:
    def myAtoi(self, s: str) -> int:
        x_str = "0"
        pos = None
        
        for i, c in enumerate(s):
            if x_str == "0" and (c == "-" or c == "+"):
                if pos != None:
                    return 0
                else:
                    pos = -1 if c == "-" else 1
                    x_str += "0"
            elif c.isnumeric():
                x_str += c
            elif c != " " or len(x_str)>1:
                break

        x = int(x_str)

        if pos == -1:
            return max(-2147483648, x*-1)
        else:
            return min(2147483647, x)
        
Solution().myAtoi("   -42")

Solution().myAtoi("1337c0d3")
```

```python
# Regular Expression Matching

class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        dp = [[False] * (len(p) + 1) for _ in range(len(s) + 1)]
        dp[0][0] = True

        for j in range(2, len(p) + 1):
            if p[j-1] == '*':
                dp[0][j] = dp[0][j-2]
                
                
        for i in range(1, len(s) + 1):
            for j in range(1, len(p) + 1):
                if p[j-1] == '.' or p[j-1] == s[i-1]:
                    dp[i][j] = dp[i-1][j-1]
                elif p[j-1] == '*':
                    dp[i][j] = dp[i][j-2]
                    if p[j-2] == '.' or p[j-2] == s[i-1]:
                        dp[i][j] |= dp[i-1][j]

        return dp[len(s)][len(p)]

        
    
Solution().isMatch("aab", "c*a*b")
```

```python
# Container With Most Water

class Solution:
    def maxArea(self, height: list[int]) -> int:
        i = 0                   # left pointer
        j = len(height)-1       # right pointer
        maxx = 0

        while True:

            # Calculate water at given step
            l = height[i]
            r = height[j]
            if l >= r: # If l is greater, then max is r
                temp = r * (j-i)
            else:
                temp = l * (j-i)

            # Check if its greater than max
            if temp > maxx:
                maxx = temp

            # Move the pointer
            if l >= r: # if the height at j is smaller, move j
                j = j - 1
            else: # else move i
                i = i + 1

            if i == j:
                return maxx
```
